```python
#Question 1
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#1
def make_sine_wave(x, A, w):
    return A* np.sin(w*x)
#2
x = np.arange(0, 2*np.pi, (0.001)*(2*np.pi))
print(x.size)
#3
Amp = np.array([0.5, 1, 1.5, 2, 2.5])
wave = np.array([1, 2, 3, 4, 5])
```

```python
#4
plt.figure(figsize = (10, 6))
#5
for A, w in zip(Amp, wave):
    Y = make_sine_wave(x, A, w)
    plt.plot(x, Y, label=f"A={A}, w={w}")
#6
plt.title("Sines!", size=50)
plt.xlabel("Dependent variable of sine waves!")
plt.ylabel("Sine Waves!")
plt.grid(True)
plt.legend()
plt.show()
```
